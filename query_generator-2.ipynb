```python
import random
import os
from pydub import AudioSegment

def generate_random_mp3_query(song_path, duration_sec, output_path="query_let_it_be.mp3"):
    """
    Extracts a random snippet from an MP3 and saves it as a new MP3.
    Requires 'pydub' and 'ffmpeg' installed on your system.
    """
    print(f"Loading: {os.path.basename(song_path)}")
    
    # Load the full audio file
    audio = AudioSegment.from_mp3(song_path)
    
    # pydub works in milliseconds
    duration_ms = int(duration_sec * 1000)
    total_duration_ms = len(audio)
    
    if total_duration_ms <= duration_ms:
        start_ms = 0
        end_ms = total_duration_ms
    else:
        start_ms = random.randint(0, total_duration_ms - duration_ms)
        end_ms = start_ms + duration_ms
        
    print(f"Extracting {duration_sec}s clip starting at {start_ms/1000:.2f}s...")
    
    # Slice the audio
    snippet = audio[start_ms:end_ms]
    
    # Export as a new MP3
    snippet.export(output_path, format="mp3")
    print(f"Success! Random query saved to: {output_path}")
    
    return output_path
```

```python
# Run this once to create your test clip
my_query_file = generate_random_mp3_query("songs/Let It Be.mp3", duration_sec=15.0)
```
